```python
import tkinter as tk
from tkinter import filedialog, messagebox
import shutil
import os
import subprocess
from IPython.display import display, Image

class FileUploaderApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Object Detector")
        
        # Create a label for file upload
        self.upload_label = tk.Label(root, text="Upload File:")
        self.upload_label.pack(pady=20)
        
        # Create a button to upload file
        self.upload_button = tk.Button(root, text="Browse Files", command=self.browse_files)
        self.upload_button.pack(pady=20)
        
        # Create a label for destination path
        self.dest_label = tk.Label(root, text="Destination Path:")
        self.dest_label.pack(pady=20)
        
        # Create an entry widget to input destination path
        self.dest_path_entry = tk.Entry(root, width=50)
        self.dest_path_entry.pack(pady=20)
        
        # Create a button to save file
        self.save_button = tk.Button(root, text="Detect Objects", command=self.save_file)
        self.save_button.pack(pady=20)
        
    def browse_files(self):
        # Open file dialog to select file
        file_path = filedialog.askopenfilename()
        
        # Update entry widget with selected file path
        self.dest_path_entry.delete(0, tk.END)
        self.dest_path_entry.insert(0, file_path)
        
    def save_file(self):
        # Get source file path
        src_file_path = self.dest_path_entry.get()
        
        # Specify destination directory path
        dest_dir_path = "C:\\Users\\MudraboinaSowryasree\\myenv\\webscraper\\webscraper\\yolov7\\inference\\images"
        
        # Create destination directory if it does not exist
        if not os.path.exists(dest_dir_path):
            os.makedirs(dest_dir_path)
        
        # Check if source file path is valid
        if src_file_path and os.path.exists(src_file_path):
            # Set destination file path
            dest_file_path = os.path.join(dest_dir_path, os.path.basename(src_file_path))
            
            # Copy file to destination directory
            shutil.copy(src_file_path, dest_file_path)
            
            # Construct the command
            command = ["python", "detect.py", "--weights", "yolov7.pt", "--conf", "0.25", "--img-size", "640", "--source", dest_file_path]
            
            try:
                # Execute the command and capture the output
                result = subprocess.run(command, capture_output=True, text=True)
                
                # Print the command output for debugging
                print("Command Output:", result.stdout)
                
                # Extract the path to the saved image from the output
                output_lines = result.stdout.split('\n')
                saved_image_path = None
                for line in output_lines:
                    if 'The image with the result is saved in:' in line:
                        saved_image_path = line.split('The image with the result is saved in:')[1].strip()
                        break
                
                if saved_image_path:
                    # Display the image from the extracted path
                    display(Image(filename=saved_image_path, width=400, height=300))
                    
                    # Show success message
                    messagebox.showinfo("Success", "Objects are detected Successfully!")
                else:
                    # Show error message if saved image path is not found
                    messagebox.showerror("Error", "Saved image path not found in output!")
            
            except Exception as e:
                # Show error message for any exceptions during command execution
                messagebox.showerror("Error", f"An error occurred: {str(e)}")
        else:
            # Show error message for invalid source file path
            messagebox.showerror("Error", "Invalid file path!")

if __name__ == "__main__":
    # Create Tkinter root window
    root = tk.Tk()
    
    # Create FileUploaderApp instance
    app = FileUploaderApp(root)
    
    # Run Tkinter event loop
    root.mainloop()
```
